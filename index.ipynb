```python
import pandas as pd
```

```python
df = pd.read_csv('Datasets/customers.csv')
df
```

```python
# Split dates into y,m,d and hour
df.info()
```

```python
df['registered_at'] = pd.to_datetime(df['registered_at'])
```

```python
df.info()
```

```python
df['registered_year'] = pd.DatetimeIndex(df['registered_at']).year
df['registered_month'] = pd.DatetimeIndex(df['registered_at']).month
df['registered_day'] = pd.DatetimeIndex(df['registered_at']).day
```

```python
# Ill slice hour from datetime string format
```

```python
df['ConvertedDate']=df['registered_at'].astype(str)
```

```python
def hourSlicer(datetime_text):
    hour = datetime_text[11:13]
    return hour
```

```python
df['registered_hour'] = df['ConvertedDate'].apply(hourSlicer)
```

```python
df
```

```python
df['registered_hour'] = df['registered_hour'].astype(int)
```

```python
df
```

```python
df = df[['id', 'name', 'email', 'registered_at', 'registered_year', 'registered_month', 'registered_day', 'registered_hour']]
```

```python
df
```

```python
df.to_csv('transformed_Datasets/customers.csv', index=False)
```
